```python
# Since the dataset is very large, let's make it a bit smaller by filtering it
# We are going to show data for 50km or 50mi races, year 2020 and country set as USA

#df_50_2020_USA = df[df["Event distance/length"].isin(["50km","50mi"]) & df["Year of event"] == "2020"]

df_50_2020_USA = df[(df["Event distance/length"].isin(["50km", "50mi"])) & (df["Year of event"] == 2020) & (df["Event name"].str.contains('USA', na=False))]

df_50_2020_USA
```

```python
# To make sure everything is correct, let's make again a value count

filtered_data = df_50_2020_USA[["Event distance/length","Event name","Year of event"]]

for count_filter in filtered_data.columns.values.tolist():
    print(df_50_2020_USA[count_filter].value_counts())
    print("")
```

```python
# Our filter is correct since the data displayed show 50km and 50mi, year 2020 and USA in the event name
```

```python
#Now that we know our filters are correct, we are going to remove "USA" from the event name

df_50_2020_USA["Event name"] = df_50_2020_USA["Event name"].str.replace("(USA)","")

df_50_2020_USA.head()
```

```python
# Now, we are going to add another column for the age of the athlete where we will basically subtract the event date from the year of birth
```

```python
#Since missing values in the "Athlete year of birth" column cannot be converted to integer, I need to drop them first from the dataset
#Then, I can convert the columns "Year of event" and "Athlete year of birth" to to integer
#Finally, I can create the "Athlete_Age" column by subtracting 'Year of event' and 'Athlete year of birth'

df_50_2020_USA = df_50_2020_USA.dropna(subset=['Athlete year of birth'])

df_50_2020_USA["Year of event"] = df_50_2020_USA["Year of event"].astype("int64")
df_50_2020_USA["Athlete year of birth"] = df_50_2020_USA["Athlete year of birth"].astype("int64")

df_50_2020_USA['Athlete_Age'] = df_50_2020_USA['Year of event'] - df_50_2020_USA['Athlete year of birth']

df_50_2020_USA[['Athlete_Age']]
```

```python
# "Athlete_Age" has been added as the last column
df_50_2020_USA.head(10)
```

```python
# Now I am going to remove the "h" from the column "Athlete performance"
df_50_2020_USA["Athlete performance	"] = df_50_2020_USA["Athlete performance"].str.replace("h","")
df_50_2020_USA = df_50_2020_USA.drop(df_50_2020_USA.columns[13], axis=1)
df_50_2020_USA['Athlete_Age'] = df_50_2020_USA['Year of event'] - df_50_2020_USA['Athlete year of birth']
```

```python
# We can see that now the "h" from "Athlete performance" column is gone 
df_50_2020_USA.head(10)
```

```python
# We are now going to drop the "Athlete club" column since it contains a lot of missing values and is not relevant as well

print("The column Athlete club contains",df_50_2020_USA["Athlete club"].isna().sum(),"missing values")
```

```python
df_50_2020_USA.drop(columns=["Athlete club"], inplace = True)
df_50_2020_USA.head(5)
```

```python
# We are also going to drop Athlete year of birth and Athlete age category

df_50_2020_USA.drop(columns=["Athlete year of birth", "Athlete age category"], inplace=True)
df_50_2020_USA.head()
```

```python
# We are going to drop the index in our dataset so that it starts from 0

df_50_2020_USA = df_50_2020_USA.reset_index(drop=True)
df_50_2020_USA.head(10)
```

```python
# There are still a few issues with the data types, so we are going to fix this

df_50_2020_USA["Athlete performance"] = pd.to_timedelta(df_50_2020_USA["Athlete performance"])
df_50_2020_USA["Athlete average speed"] = df_50_2020_USA["Athlete average speed"].astype(float)

print(df_50_2020_USA.dtypes)
```

```python
# Since I have datetime with the following format : "28.-29.11.2020", I need to write a function to take the first value and then convert it to datetime with the european format

# Function to extract the first date

import re

# Function to reformat dates
def reformat_date(date_str):
    # Check if the date contains a range
    if '-' in date_str:
        # Extract the first date part of the range
        match = re.match(r'(\d{2})\.-\d{2}\.(\d{2}\.\d{4})', date_str)
        if match:
            # Construct the reformatted date string
            first_date = f"{match.group(1)}.{match.group(2)}"
            return first_date
    return date_str

# Apply the function to the "Event dates" column
df_50_2020_USA['Event dates'] = df_50_2020_USA['Event dates'].apply(reformat_date)

# Display the DataFrame after reformatting
print("DataFrame after reformatting 'Event dates':")
print(df_50_2020_USA)
```

```python
# When trying to convert the date to datetime, another error popped up because it exists some dates with the following format "29.02.-01.03.2020"
# Therefore, I will simply replace it by "29.02.2020"

df_50_2020_USA['Event dates'] = df_50_2020_USA['Event dates'].replace('29.02.-01.03.2020', '29.02.2020')
```

```python
# Convert "Event dates" column to datetime with dayfirst parameter

df_50_2020_USA['Event dates'] = pd.to_datetime(df_50_2020_USA['Event dates'], dayfirst=True)
```

```python
# Plot the histogram for Average Speed During 50km races
plt.figure(figsize=(10, 6))
sns.displot(df_50_2020_USA[df_50_2020_USA["Event_Distance"] == "50km"]['Athlete_Avg_Speed'])
plt.xlabel('Average Speed (km/h)')
plt.ylabel('Frequency')
plt.title('Distribution of Athlete Average Speed During 50km races')
plt.show()
```

```python
# First, we can see that the average speed is distributed normally, with a bell-shape curve. 
# The average speed for athletes tend to be around 8km/h
# There are very few athletes with average speeds below 4 km/h or above 14 km/h. These can be considered outliers, as they fall far from the central mass of the distribution.
```

```python
# Athlete Average Speed by Gender

df_50km = df_50_2020_USA[df_50_2020_USA['Event_Distance'] == '50km']

plt.figure(figsize=(10, 6))
sns.boxplot(x='Athlete_Gender', y=df_50km['Athlete_Avg_Speed'], data=df_50_2020_USA)
plt.xlabel("Gender")
plt.ylabel("Average Speed (km/h)")
plt.title('Distribution of Athletes Average Speed by Gender for 50km races')
plt.show()
```

```python
# The median for both Male and Female is quite similar, indicating a slighty higher average speed for Men
# since the Interquartile Range for both Male and Female is similar, The middle 50% of athletes have similar average speed, ranging from 6km
# The presence of outliers indicate variability in performance. Men tend to perfom better when it comes to top speed ranges
```

```python
# another way to display the distribution of average speed by gender is to use a violon plot
# We can add a variable in the graph such as the distance of the event

sns.violinplot(x="Athlete_Gender", y="Athlete_Avg_Speed", data=df_50_2020_USA, hue="Event_Distance", split=True, inner = "quart", linewidth = 1)
```

```python
# Average Speed vs Age

plt.figure(figsize=(10, 6))
sns.regplot(x='Athlete_Age', y='Athlete_Avg_Speed', data=df_50km, scatter_kws={'s':10})
plt.xlabel("Athletes Age")
plt.ylabel('Average Speed (km/h)')
plt.title('Average Speed vs Athletes Age')
plt.show()
```

```python
# although there are variability across age groups, there is a noticable decrease in average speed as age increase, particularly after 50 years old
# Athletes aged between 20 and 40 years old tend to have the highest average speed. This is where outliers are located
```
